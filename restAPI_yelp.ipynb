```python
def get_address(x):
    if x['address2'] == str:
        address = x['address1'] + ' ' + x['address2'] + ', ' + x['city'] + ' ' + x['state'] + ' ' + x['zip_code']
    else:
        address = x['address1'] + ' ' + x['city'] + ' ' + x['state'] + ' ' + x['zip_code']
    return address
```

```python
df['address'] = df['location'].apply(get_address)
```

```python
def get_zipcode(x):
    zip_code = x['zip_code']
    return zip_code
```

```python
df['zip_code'] = df['location'].apply(lambda x: x['zip_code'])
```

```python
def get_cat_list(x):
    cat_list = []
    n = len(x)
    for i in range(n):
        cat = x[i]['title']
        cat_list.append(cat)
    return cat_list
```

```python
df['cat_list'] = df['categories'].apply(get_cat_list)
```
